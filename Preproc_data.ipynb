```python
# Import librairies:
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Financial Statement documents:

```python
# Balance Sheet:
balance_sheet=pd.read_excel(r'Financial Statement(SID).xlsx',sheet_name= 'Balance Sheet annualy')
balance_sheet=balance_sheet.iloc[2:,1:-1].reset_index(drop=True)

# Rename columns:

balance_sheet.columns=balance_sheet.columns.str.replace('FY','',regex=False)
balance_sheet
```

```python
# Income Statement:

income_statement=pd.read_excel(r'Financial Statement(SID).xlsx',sheet_name='Income Statement annualy')
income_statement=income_statement.iloc[2:,1:-1].reset_index(drop=True)

# Rename columns:

income_statement.columns=income_statement.columns.str.replace('FY','',regex=False)
income_statement
```

```python
# Cash Flow Statement:
cash_flow_statement=pd.read_excel(r'Financial Statement(SID).xlsx',sheet_name='Cash Flow annualy')
cash_flow_statement=cash_flow_statement.iloc[2:,1:-1].reset_index(drop=True)

# Rename columns:

cash_flow_statement.columns=cash_flow_statement.columns.str.replace('FY','',regex=False)
cash_flow_statement
```

```python
ratios_statement=pd.read_excel(r'Financial Statement(SID).xlsx',sheet_name='Ratios annualy')
ratios_statement=ratios_statement.iloc[2:,1:-1].reset_index(drop=True)
ratios_statement.columns=ratios_statement.columns.str.replace('FY','',regex=False)
ratios_statement
```

### Transpose the Data :

##### Balance Sheet:

```python
# Balance Sheet:
bs_trsp=balance_sheet.transpose()
bs_trsp=bs_trsp.reset_index(drop=False)
# sets first row as the actual column names:
bs_trsp.columns=bs_trsp.iloc[0]
# removes row 0:
bs_trsp=bs_trsp.drop(0)
bs_trsp
```

```python
# Convert to datetime:
bs_trsp[bs_trsp.columns[0]] = pd.to_datetime(bs_trsp.iloc[:, 0], errors='coerce')
# Set the date to December 31 of the year for all valid dates:
bs_trsp[bs_trsp.columns[0]] = bs_trsp[bs_trsp.columns[0]].dt.year.astype('int64').astype(str) + '-12-31'
bs_trsp[bs_trsp.columns[0]]=pd.to_datetime(bs_trsp[bs_trsp.columns[0]],errors='coerce')
bs_trsp
```

##### Income Statement:

```python
inc_trsp=income_statement.transpose().reset_index(drop=False)
# sets first row as the actual column names:
inc_trsp.columns=inc_trsp.iloc[0]
# Remove the row 0:
inc_trsp=inc_trsp.drop(0)
inc_trsp
```

```python
#Change data type of Fiscal Year:

inc_trsp['Fiscal Year']=pd.to_datetime(inc_trsp['Fiscal Year'], errors='coerce')
# Set the date to December 31 of the year for all valid dates:
inc_trsp['Fiscal Year']=inc_trsp['Fiscal Year'].dt.year.astype('int64').astype(str)+'-12-31'
inc_trsp['Fiscal Year']=pd.to_datetime(inc_trsp['Fiscal Year'],errors='coerce')
inc_trsp
```

##### Cash Flow Statement:

```python
cf_trsp=cash_flow_statement.transpose().reset_index(drop=False)
# sets first row as the actual column names:
cf_trsp.columns=cf_trsp.iloc[0]
# Remove first row:
cf_trsp=cf_trsp.drop(0)
cf_trsp
```

```python
# Change data type of fiscal year:
cf_trsp['Fiscal Year']=pd.to_datetime(cf_trsp['Fiscal Year'], errors='coerce')
# Set the date to December 31 of the year for all valid dates:
cf_trsp['Fiscal Year']=cf_trsp['Fiscal Year'].dt.year.astype(str)+'-12-31'
# Convert the resulting strings back to datetime:
cf_trsp['Fiscal Year']=pd.to_datetime(cf_trsp['Fiscal Year'],errors='coerce')
cf_trsp
```

```python
rat_trsp=ratios_statement.transpose().reset_index(drop=False)
rat_trsp.columns=rat_trsp.iloc[0]
rat_trsp=rat_trsp.drop([0,1]).reset_index(drop=True)
# Set new index starting from 1
rat_trsp.index = rat_trsp.index + 1
rat_trsp
```

```python
rat_trsp['Fiscal Year']=pd.to_datetime(rat_trsp['Fiscal Year'],errors='coerce')
rat_trsp['Fiscal Year']=rat_trsp['Fiscal Year'].dt.year.astype(int).astype(str)+'-12-31'
rat_trsp['Fiscal Year']=pd.to_datetime(rat_trsp['Fiscal Year'],errors='coerce')
rat_trsp
```

### Preprocessing:

```python
# Create new variable:
 ## activity ratios:
revenue=inc_trsp['Revenue'].str.replace(',','').astype(float)
cogs=inc_trsp['Cost of Revenue'].str.replace(',','').astype(float)
receivable=bs_trsp['Receivables'].str.replace(',','').astype(float)
inventory=bs_trsp['Inventory'].str.replace(',','').astype(float)
payable=bs_trsp['Accounts Payable'].str.replace(',','').astype(float)
current_assets=bs_trsp['Total Current Assets'].str.replace(',','').astype(float)
current_liabilities=bs_trsp['Total Current Liabilities'].str.replace(',','').astype(float)
total_assets=bs_trsp['Total Assets'].str.replace(',','').astype(float)
total_liabilities=bs_trsp['Total Liabilities'].str.replace(',','').astype(float)
total_equity=bs_trsp['Shareholders\' Equity'].str.replace(',','').astype(float)
 ##liquidity ratios:
cash_equivalents=bs_trsp['Cash & Equivalents'].str.replace(',','').astype(float)
operating_cash_flow=cf_trsp['Operating Cash Flow'].str.replace(',','').astype(float)
investing_cash_flow=cf_trsp['Investing Cash Flow'].str.replace(',','').astype(float)
financing_cash_flow=cf_trsp['Financing Cash Flow'].str.replace(',','').astype(float)
net_cash_flow=cf_trsp['Net Cash Flow'].str.replace(',','').astype(float)
 ##Solvency ratios:
total_debt=bs_trsp['Total Debt'].str.replace(',','').astype(float)
ebit=inc_trsp['EBIT'].str.replace(',','').astype(float)
interest_expenses=inc_trsp['Interest Expense'].str.replace(',','').astype(float).abs() #abs() is absolute number
 ##Profitability ratios:
gross_profit=inc_trsp['Gross Profit'].str.replace(',','').astype(float)
ebt=inc_trsp['Pretax Income'].str.replace(',','').astype(float)
net_income=inc_trsp['Net Income'].str.replace(',','').astype(float)
roce = rat_trsp['Return on Capital Employed (ROCE)'].str.replace('%', '',regex=False).replace('-',np.nan).astype(float) / 100
roic=rat_trsp['Return on Capital (ROIC)'].str.replace(',','').str.replace('%','').replace('-',np.nan).astype(float)/100
 ##Valuation ratios:
eps=inc_trsp['EPS (Basic)'].str.replace(',','').astype(float)
pe=rat_trsp['PE Ratio'].str.replace(',','').replace('-',np.nan).astype(float)
pb=rat_trsp['PB Ratio'].str.replace(',','').replace('-',np.nan).astype(float)
ps=rat_trsp['PS Ratio'].str.replace(',','').replace('-',np.nan).astype(float)
ev_ebitda=rat_trsp['EV/EBITDA Ratio'].str.replace(',','').astype(float)
dividend_yield=rat_trsp['Dividend Yield'].str.replace(',','').str.replace('%','').astype(float)/100



```

#### 0- Overview:

```python
overview=inc_trsp[['Fiscal Year']].copy()
overview['Revenue']=revenue
overview['COGS']=cogs
overview['Gross Profit']=gross_profit
overview['Operating Income(EBIT)']=ebit
overview['Net Income']=net_income
overview['Total Assets']=total_assets
overview['Total liabilities']=total_liabilities
overview['Current Assets']=current_assets
overview['Current Liabilities']=current_liabilities
overview['Operating CF']=operating_cash_flow
overview['Investing CF']=investing_cash_flow
overview['Financing CF']=financing_cash_flow
overview['Net Cash Flow']=net_cash_flow

overview
```

#### 1- Activity Ratios:

```python
# Create new data frame:
act_ratio=inc_trsp[['Fiscal Year']].copy()
act_ratio['Revenue']=revenue
act_ratio['COGS']= cogs
act_ratio['DSO']=(receivable/revenue)*365
act_ratio['DIO']=(inventory/cogs)*365
act_ratio['DPO']=(payable/cogs)*365
act_ratio['CCC']=act_ratio['DSO']+act_ratio['DIO']-act_ratio['DPO']
act_ratio['Working Capital']=current_assets - current_liabilities
act_ratio['Beginning Working Capital']=act_ratio['Working Capital'].shift(-1,fill_value=act_ratio['Working Capital'].iloc[-1])
act_ratio['Working Capital Turnover']=revenue/((act_ratio['Working Capital']+act_ratio['Beginning Working Capital'])/2)
act_ratio['Total Assets']=total_assets
act_ratio['Beginning Total Assets']=act_ratio['Total Assets'].shift(-1,fill_value=act_ratio['Total Assets'].iloc[-1])
act_ratio['Total Assets Turnover']=revenue/((act_ratio['Total Assets']+act_ratio['Beginning Total Assets'])/2)
act_ratio['Total Equity']=total_equity
act_ratio['Beginning Total Equity']=act_ratio['Total Equity'].shift(-1,fill_value=act_ratio['Total Equity'].iloc[-1])
act_ratio['Equity Turnover']=revenue/((act_ratio['Total Equity']+act_ratio['Beginning Total Equity'])/2)
act_ratio
```

#### 2- Liquidity Ratios:

```python
liq_ratio=bs_trsp[['Fiscal Year']].copy()
liq_ratio['Current Ratio']=current_assets/current_liabilities
liq_ratio['Quick Ratio']=(current_assets-inventory)/current_liabilities
liq_ratio['Cash Ratio']=cash_equivalents/current_liabilities
liq_ratio['Operating Cash Flow']=operating_cash_flow/current_liabilities
liq_ratio
```

#### 3- Solvency Ratios:

```python
solv_ratios=bs_trsp[['Fiscal Year']].copy()
solv_ratios['Debt to Equity (D/E)']=total_debt/total_equity
solv_ratios['Debt to Capital']=total_debt/(total_debt+total_equity)
solv_ratios['Debt to Assets']=total_debt/total_assets
solv_ratios['Finanicial Leverage']= total_assets /total_equity
solv_ratios['Interest Coverage']=ebit/interest_expenses
solv_ratios
```

#### 4- Profitability Ratios:

```python
prof_ratios=bs_trsp[['Fiscal Year']].copy()
prof_ratios['Gross Profit Margin']=gross_profit/revenue
prof_ratios['Operating Profit Margin']=ebit/revenue
prof_ratios['Pre-tax Margin']=ebt/revenue
prof_ratios['Net Profit Margin']=net_income/revenue
prof_ratios['Return on Assets(ROA)']= net_income/total_assets
prof_ratios['Return on Equity(ROE)']=net_income/total_equity
prof_ratios['Return on Capital Employed(ROCE)']=roce
prof_ratios['Return on Invested Capital(ROIC)']=roic
prof_ratios
```

#### 5- Valuation Ratios:

```python
val_ratios=inc_trsp[['Fiscal Year']].copy()
val_ratios['EPS']=eps
val_ratios['Price to Earnings(P/E)']=pe
val_ratios['Price to Sales(P/S)']=ps
val_ratios['Price to Book(P/B)']=pb
val_ratios['EV/EBITDA']=ev_ebitda
val_ratios['Dividend Yield']=dividend_yield
val_ratios
```

```python
data={
    'Balance Sheet':bs_trsp,
    'Income Statement':inc_trsp,
    'Cash Flow':cf_trsp,
    'Sonasid Overview':overview,
    'Sonasid Activity':act_ratio,
    'Sonasid Liquidity':liq_ratio,
    'Sonsid Solvency':solv_ratios,
    'Sonasid Profitability':prof_ratios,
    'Sonasid Valuation':val_ratios
}
file_name='Financial Ratios (Sonasid).xlsx'
with pd.ExcelWriter(file_name,engine='xlsxwriter') as writer:
    for name,df in data.items():
        df.to_excel(writer,sheet_name=name,index=False)
print(f"Excel File '{file_name}'saved successfully")
```
